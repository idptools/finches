### Compute Heterotypic Epsilon vectors with surface of FOLDED DOMAIN

```python
nterminal_IDR = 'MSDNGPQNQRNAPRITFGGPSDSTGSNQNGERSGARSKQRRPQGLPNNT'
PDB_FILE = 'test_pdb.pdb'

# initialize forcefild model
mPiPi_GGv1_model = mPiPi_model('mPiPi_GGv1')
CALVADOS2_model = calvados_model('CALVADOS2')

# intialize matrix constructor with model
X_local = epsilon_calculation.Interaction_Matrix_Constructor(mPiPi_GGv1_model) 


#  NOTE: domain bounds IDR: 0-49, FD: 49-173
# parse folded domain
SAFD_seq, SAFD_idxs, SAFD_cords = PDB_structure_tools.pdb_to_SDFDresidues_and_xyzs(PDB_FILE,  49, 173, issolate_domain=False)

# compute epsilon
print(f'IDR Centric Epsilon Vector:')
defult_attractive_vector, defult_repulsive_vector = epsilon_calculation.get_interdomain_epsilon_vectors(SAFD_seq, nterminal_IDR, X_local, SAFD_cords, IDR_positon='Nterm')
print(f' Epsilon calculation (Default): {defult_attractive_vector, defult_repulsive_vector}')

NOCHARGE_attractive_vector, NOCHARGE_repulsive_vector  = epsilon_calculation.get_interdomain_epsilon_vectors(SAFD_seq, nterminal_IDR, X_local, SAFD_cords, IDR_positon='Nterm', CHARGE=False)
# print(f' Epsilon calculation (NO charge weighting): {NOCHARGE_attractive_vector, NOCHARGE_repulsive_vector }')

# compute epsilon
print(f'\nFD Centric Epsilon Vector:')
defult_calcuation = epsilon_calculation.get_interdomain_epsilon_vectors(SAFD_seq, nterminal_IDR, X_local, SAFD_cords, IDR_positon='Nterm', sequence_of_ref='sequence2')
print(f' Epsilon calculation (Default): {defult_attractive_vector, defult_repulsive_vector}')

NOCHARGE_attractive_vector, NOCHARGE_repulsive_vector = epsilon_calculation.get_interdomain_epsilon_vectors(SAFD_seq, nterminal_IDR, X_local, SAFD_cords, IDR_positon='Nterm', CHARGE=False,  sequence_of_ref='sequence2')
# print(f' Epsilon calculation (NO charge weighting): {NOCHARGE_attractive_vector, NOCHARGE_repulsive_vector }')
```
