## Encode them into interaction vectors

```python
rg_sz = 9
std=1.8

wt_enc = mf.calculate_region_interactions_mat(BARD1_wt, region_size=rg_sz, kernal_type='gaussian', std=std)
inc_enc = mf.calculate_region_interactions_mat(BARD1_inc, region_size=rg_sz, kernal_type='gaussian', std=std)
```

```python
# calc the dtw
cost, mat = calculate_dtw_distance_flex(wt_enc, inc_enc, step_pattern='symmetric2')
path = find_dtw_path_flex(mat, step_pattern='symmetric2')
```
